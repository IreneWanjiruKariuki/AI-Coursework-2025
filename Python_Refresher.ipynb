QUESTION 1

```python
# Ask the user for the number of days
days = int(input("Enter the number of days: "))

# Calculate the number of seconds in the given days
seconds = days * 24 * 60 * 60

# Display the result
print(f"There are {seconds} seconds in {days} days.")
```

QUESTION 2

```python
import math

# Ask the user to input the radius
radius = float(input("Enter the radius of the sphere: "))

# Calculate the volume
volume = (4/3) * math.pi * (radius ** 3)

# Display the result
print(f"The volume of the sphere with radius {radius} is {volume:.2f}")
```

QUESTION 3

```python
# Function to calculate the area of a square
def calculate_area(side):
    return side ** 2

# Function to calculate the perimeter of a square
def calculate_perimeter(side):
    return 4 * side

# Enter the side length of the square
side_length = float(input("Enter the length of the side of the square: "))

# Call the functions
area = calculate_area(side_length)
perimeter = calculate_perimeter(side_length)

# Display the results
print(f"The area of the square is: {area}")
print(f"The perimeter of the square is: {perimeter}")
```

QUESTION 4

```python
# Function to check whether a character is uppercase or lowercase
def check_case(character):
    if character.isupper():
        print(f"The character '{character}' is UPPERCASE.")
    elif character.islower():
        print(f"The character '{character}' is lowercase.")

# Ask the user to input a character
char = input("Enter a character: ")

check_case(char)
```

QUESTION 5

```python
# Initialize variables
x = 0
y = 20

# Repeat until y is less than 6
while y >= 6:
    y = y - 4
    x = x + (2 / y)

# Display the result
print("The final value of x is:", x)
```

QUESTION 6

```python
# Initialize an empty list to store the values
values = []

# Use a loop to collect 5 values from the user
for i in range(5):
    num = float(input(f"Enter value {i + 1}: "))
    values.append(num)

# Calculate the average
average = sum(values) / len(values)

# Display the results
print(f"The average of the entered values is: {average}")
```
